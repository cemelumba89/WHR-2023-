# 03. Creating a Correlation Matrix Heatmap (colored)

```python
#### Create a correlation matrix using pandas
df_whrsub.corr()
```

```python
#### Create a correlation heatmap using matplotlib

plt.matshow(df_whrsub.corr())
plt.show()
```

```python
#### Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_whrsub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_whrsub.shape[1]), df_whrsub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_whrsub.shape[1]), df_whrsub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title

# Save the plot
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Correlation Matrix.png'))
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix above.
corr = sns.heatmap(df_whrsub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

### Comments: Since all correlation values are not appearing on the heatmap and this may be due to them being very close to zero or equal to zero, I will adjust 'fmt' parameter. This adjustment of the format specifier (fmt) should help ensure that even small or zero correlation values are displayed on the heatmap.

```python
#### Recreation of visualization with adjustment of 'fmt' parameter
# Create a subplot with matplotlib
f, ax = plt.subplots(figsize=(10, 10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix above.
corr = sns.heatmap(df_whrsub.corr(), annot=True, fmt=".2f", ax=ax)
```

# 05. Creating Categorical Plot

```python
##### Creating a categorical variable that splits the "Life Ladder" column into categories. Use a histogram to visualize the distribution of the variables. 

sns.histplot(df_whrclean['Life Ladder'], bins = 20, kde = True)

# Save the plot
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'histplot.png'))
```

```python
df_whrclean.loc[df_whrclean['Life Ladder'] < 4, 'Ladder score'] = 'Low score'
```

```python
df_whrclean.loc[(df_whrclean['Life Ladder'] >= 4) & (df_whrclean['Life Ladder'] < 6), 'Ladder score'] = 'Middle score'
```

```python
df_whrclean.loc[df_whrclean['Life Ladder'] >= 6, 'Ladder score'] = 'High score'
```

```python
df_whrclean['Ladder score'].value_counts(dropna = False)
```

```python
df_whrclean.columns
```

```python
#### Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="Log GDP per capita", y="Life Ladder", hue="Ladder score", data=df_whrclean)

# Save the plot
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'catplot.png'))
```

#### Comments: This catplot demonstrates that most countries are represented by a Ladder score that would be considered a Middle score. This is consistent with the value_count executed on df_whrclean. Thankfully, there are only a few countries with a Ladder Score that would be considered Low. This is particularly interesting because from the Statistical Appendix of the WHR 2023, the mean Ladder score for the entire dataset is 5.48.
